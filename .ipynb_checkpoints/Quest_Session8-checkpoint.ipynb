```python
import pandas as pd
import numpy as np

data = pd.readㅠ_csv('session8_dataset/2017.csv')
data.head()
```

```python
data.info()
```

```python
# dataframe에 isnull()을 적용하면 각 원소가 null인지 True/False 여부로 바꾼 dataframe 반환
# 여기에 sum()을 실행하면 각 데이터프레임의 칼럼별로 value를 합산하기 때문에 결국 true의 개수를 알 수 있다
data.isnull().sum()
```

```python
from sklearn import preprocessing
scaler = preprocessing.StandardScaler() 

data_scaled = scaler.fit_transform(data.iloc[:,2:])
data_scaled = pd.DataFrame(data_scaled, columns=data.columns[2:])
data_scaled.head()
```

```python
import matplotlib.pyplot as plt
# kde plot for all columns in dataframe
# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.kde.html
# subplot으로 넣는 방법 찾아보기
ax1 = data.iloc[:,2:].plot.kde()
ax2 = data_scaled.plot.kde()
ax1.set_title('Before Scaling')
ax2.set_title('After Scaling')
plt.show()
```

```python
# column 이름으로 df의 일부 고르기 --> .loc[모든 row, 칼럼 이름(들의 리스트)]
# df.values : 데이터 프레임을 하나의 데이터 포인트에 대한 [values for each attribute]의 리스트로 변환
X = data.loc[:,['Health..Life.Expectancy.', 'Freedom']].values
```

```python
from sklearn.cluster import KMeans

# 적절한 k 값 찾기
score=[]
for i in range(1,11): # 아무리 많아도 10까지만 검증하면 된다.
    kmeans = KMeans(n_clusters= i, init='k-means++', random_state=0)
    kmeans.fit(X)
    score.append(kmeans.inertia_) # elbow graph의 score 값
```

```python
#Visualizing the ELBOW method to get the optimal value of K 
plt.figure(figsize=(8, 8))
plt.plot(range(1,11), score)
plt.title('The Elbow Method')
plt.xlabel('no of clusters')
plt.ylabel('wcss')
plt.show() # 5에서 마지막으로 꺾인다
```

```python
#Model Build
kmeansmodel = KMeans(n_clusters= 5, init='k-means++', random_state=0)
y_kmeans= kmeansmodel.fit_predict(X) # unsupervised에서는 fit_transform 대신 fit_predict
```

```python
# 155개의 datapoint에 최종적으로 할당된 cluster의 번호를 담은 array
y_kmeans
```
