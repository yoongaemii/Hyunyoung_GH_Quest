# Session3. 빅데이터와 머신러닝
Scipy 라이브러리로 2차원 cost function을 최소화하는 값 찾기  

```python
import scipy.optimize as op
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d, Axes3D
import numpy as np
```

```python
fun = lambda x: (1-x[0])**2 + 100*(x[1]-x[0]**2)**2
```

## 그래프 그리기  
http://jakevdp.github.io/mpl_tutorial/tutorial_pages/tut5.html  
https://stackoverflow.com/questions/3810865/matplotlib-unknown-projection-3d-error
- x,y 축 범위는 어떻게 설정해야하지?  
- Transpose는 왜 하지?  
- 지금 모양이 짱이상한데? 왜 y 영향을 안받지?

```python
x = np.outer(np.linspace(-100, 100, 30), np.ones(30)) # 30 by 30 matrix with elements from -3 to 3
y = x.copy().T # why Transpose?
z = fun([x, y])

fig = plt.figure()
ax = Axes3D(fig)

ax.plot_surface(x, y, z, cmap=plt.cm.jet, rstride=1, cstride=1, linewidth=0)
plt.xlabel('x'); plt.ylabel('y')
```

## minimize 써보기
Scipy.optimize.minimize https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.minimize.html  

```python
x0 = [1, 2]
result = op.minimize(fun,x0)
result
```
